```python
sns.pairplot(data, vars=['field1', 'field2', 'field3', 'field4', 'field5', 'field6'], hue='field1')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Loop through all columns in the dataset
for column in data.columns:
    # Skip columns that are not categorical (or convert them if needed)
    if data[column].dtype == 'object' or data[column].dtype == 'int64' or data[column].dtype == 'float64':
        # Count the values in the current column
        value_counts = data[column].value_counts()
        
        # Plot the distribution as a bar chart if there are more than one unique value
        if len(value_counts) > 1:
            value_counts.plot(kind='bar', title=f'{column} Distribution')
            plt.show()
```

```python
# Select only numerical columns for correlation calculation
numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns

# Calculate correlation matrix
correlation = data[numerical_columns].corr()

# Plot heatmap
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
